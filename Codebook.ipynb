```python
# Fun to load the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

```python
#fun to display users  and there connections
def display_users(data):
    print("\n \n User and there pages and there connections are : \n")
    for users in data['users']:
        print(f"ID: {users['id']}. {users['name']} is friends with ID: {users['friends']} and liked pages are {users['liked_pages']}")
    print("\n \n Page Information \n")
    for pages in data['pages']:
        print(f"{pages['id']} : {pages['name']}")
        
```

```python
display_users(data)
```
